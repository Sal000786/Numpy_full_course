```python
import numpy as np
```

```python
arr=np.array([1,2,3,4,5])
print(arr)
copy=arr.copy()
print(copy)
```

```python
arr=np.array([1,2,3,4,5])
print(arr)
view=arr.view()
print(view)
```

# 


```python
arr=np.arange(10)
print("original data",arr)
co=arr.copy()
print('copied data',co)
co[1]=15
print("first element changed",co)
print('still not affecting the original data',arr)
```

```python
arr=np.arange(10)
print("original data",arr)
vi=arr.view()
print('view data',vi)
vi[1]=15
print("first element changed",vi)
print('view affects the original data',arr)
```
