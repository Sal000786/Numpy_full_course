```python
import numpy as np
```

```python
arr=np.random.randint(1,10,5)
print(arr)
print("min: ",np.min(arr))
print("max: ",np.max(arr))

print("Argmax function gives the postion of the min or max element in the array")
print("min position: ",np.argmin(arr))
print("max position: ",np.argmax(arr))

print("Sqrt function returns the squre root of each element in the array")
print("sqrt: ",np.sqrt(arr))

print("Sin and cos function returns the sin and cos value of each element in the array")
print("Sin: ",np.sin(arr))
print("Cos: ",np.cos(arr))


print("Cumsum returns the sum of i and i+1 element and again the result with the next position ")
print("Cumsum: ",np.cumsum(arr))
```
