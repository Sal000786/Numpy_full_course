# Zeros function

```python
import numpy as np
```

```python
arr=np.zeros(5)
print(arr)
```

```python
arr1=np.zeros((3,3))
print(arr1)
print(arr1.ndim)
```

```python
arr2=np.zeros(((3,3,3)))
print(arr2)
print(arr2.ndim)
```

# Ones Function

```python
arr1=np.ones(5)
print(arr1)
print(arr1.ndim)
```

```python
arr2=np.ones((5,5))
print(arr2)
print(arr2.ndim)
```

```python
arr3=np.ones(((3,3,3)))
print(arr3)
print(arr3.ndim)
```


# Empty array


```python
arr1=np.empty(5)
print(arr1)


# an Empty numpy array is not actually empty it stores the value of latest variable in the program 
#otherwise it throws some sort of garbage value
```

# Numpy array using arrange function

```python
arr=np.arange(1,5)
print(arr)
```

```python
arr2=np.arange(10,50)
print(arr2)
```

# LineSpace function

```python
arr=np.linspace(0,20,num=5)
print(arr)
```

```python
arr2=np.linspace(1,10,num=3)
print(arr2)
```
