```python
import numpy as np
```

```python
mat=np.matrix([[1,2,3],[4,5,6]])
print(mat)
print(mat.T)
print()

x=np.transpose(mat)
print(x)
print(x.T)
```

```python
mat=np.matrix([[1,2,3],[4,5,6]])
print(mat)
print()
x=np.swapaxes(mat,0,1)
print(x)
print()
x=np.swapaxes(mat,1,0)
print(x)
```

```python
mat=np.matrix([[1,2],[4,5]])
print(mat)
print(np.linalg.inv(mat))
```

```python
mat=np.matrix([[1,2],[4,5]])
print(mat)
print(np.linalg.matrix_power(mat,0))
```

```python
mat=np.matrix([[1,2],[4,5]])
print(mat)
print(np.linalg.matrix_power(mat,1))
```

```python
mat=np.matrix([[1,2],[4,5]])
print(mat)
print(np.linalg.matrix_power(mat,-1))
```

```python
mat=np.matrix([[1,2],[4,5]])
print(mat)
print(np.linalg.det(mat))
```
