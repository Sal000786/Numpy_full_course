```python
import numpy as np
```

```python
print("Shuffle function to suffle the elements of an array")
arr=np.array([1,2,3,4,5])
print(arr)
np.random.shuffle(arr)
print(arr)
```

```python
print("Numpy's unique function to eleminate the repeated values in the array")
arr=np.array([1,2,3,4,5,1,2,5,8,5,8,6,9,2,7,8,5,1,0,6,0])
print(arr)
x=np.unique(arr,return_index=True,return_counts=True)
print(x)
```

```python
print("Numpy's resize function to resize an array to a different size")
arr=np.array([1,2,3,4,5,1,2,5,8,5,8,6,9,2,7,8,5,1,0,6,0])
print(arr)
x=np.resize(arr,(8,3))
print(x)
print()
x=np.resize(arr,(2,3))
print(x)
```

```python
print("Flatten in numpy array")
arr=np.array([[1,2,3],[4,5,6]])
arr=np.resize(arr,(3,2))
print(arr)
print()
x=arr.flatten(order="F")
print("Flatten : ",x)
print()
print("Ravel : ",x.ravel(order="F"))
```

```python
print("Flatten in numpy array")
arr=np.array([[1,2,3],[4,5,6]])
arr=np.resize(arr,(3,2))
print(arr)
print()
x=arr.flatten(order="A")
print("Flatten : ",x)
print()
print("Ravel : ",x.ravel(order="A"))
```
